```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the .json file directly into a DataFrame
df = pd.read_json('data.json')

# Transpose the DataFrame to organize the data
df = df.transpose()
df.columns = ['Subject', 'Grade', 'Timestamp']

# Convert 'Timestamp' to datetime format
df['Timestamp'] = pd.to_datetime(df['Timestamp'], unit='s')

# Plot 1: Distribution of grades
plt.figure(figsize=(8, 6))
plt.hist(df['Grade'], bins=10, edgecolor='black')
plt.xlabel('Grade')
plt.ylabel('Frequency')
plt.title('Distribution of Grades')
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot 2: Mean grade per subject
mean_grades = df.groupby('Subject')['Grade'].mean()
mean_grades.plot(kind='bar', xlabel='Subject', ylabel='Mean Grade', title='Mean Grade per Subject')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plot 1: Grades evolution over time for History
geo_df = df[df['Subject'] == 'History']

# Resample the data on a monthly basis and calculate the mean grade for each month
monthly_mean_grades = geo_df.set_index('Timestamp').resample('M')['Grade'].mean()

# Plot: Grades evolution over time for the "History" subject
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(monthly_mean_grades.index, monthly_mean_grades.values, marker='o', linestyle='-')
ax.set_xlabel('Month')
ax.set_ylabel('Mean Grade')
ax.set_title('Mean Grade Evolution for History (Monthly)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot 2: Grades evolution over time for English
geo_df = df[df['Subject'] == 'English']

# Resample the data on a monthly basis and calculate the mean grade for each month
monthly_mean_grades = geo_df.set_index('Timestamp').resample('M')['Grade'].mean()

# Plot: Grades evolution over time for the "English" subject
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(monthly_mean_grades.index, monthly_mean_grades.values, marker='o', linestyle='-')
ax.set_xlabel('Month')
ax.set_ylabel('Mean Grade')
ax.set_title('Mean Grade Evolution for English (Monthly)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot 3: Grades evolution over time for Math
geo_df = df[df['Subject'] == 'Math']

# Resample the data on a monthly basis and calculate the mean grade for each month
monthly_mean_grades = geo_df.set_index('Timestamp').resample('M')['Grade'].mean()

# Plot: Grades evolution over time for the "Math" subject
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(monthly_mean_grades.index, monthly_mean_grades.values, marker='o', linestyle='-')
ax.set_xlabel('Month')
ax.set_ylabel('Mean Grade')
ax.set_title('Mean Grade Evolution for Math (Monthly)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calculate the average grade over time
average_grade_over_time = df.set_index('Timestamp')['Grade'].expanding().mean()

# Plot: Evolution of average grade over time
plt.figure(figsize=(10, 6))
plt.plot(average_grade_over_time.index, average_grade_over_time.values, marker='o', linestyle='-')
plt.xlabel('Time')
plt.ylabel('Average Grade')
plt.title('Evolution of Average Grade over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
